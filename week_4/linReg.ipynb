```python
import pandas
from sklearn.feature_extraction import DictVectorizer
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import coo_matrix, hstack
from sklearn.linear_model import Ridge
```

```python
data = pandas.read_csv('salary-train.csv')
print(data)
```

```python
#data['FullDescription'] = text(data['FullDescription']).lower()
for i in range(len(data['FullDescription'])):
    if(i % 100 == 0):
        print(i, ' < ', len(data['FullDescription']))
    data['FullDescription'][i] = data['FullDescription'][i].lower()
#LocNorm =np.array(data['LocationNormalized'])
#conT = np.array(data['ContractTime'])
#salarNorm = np.array(data['SalaryNormalized'])
print(data)
```

```python
train = data
train['FullDescription'] = train['FullDescription'].replace('[^a-zA-Z0-9]', ' ', regex = True)
```

```python
full = TfidfVectorizer(min_df=5)
text = full.fit_transform(train["FullDescription"])
print(text.shape)
```

```python
train['LocationNormalized'].fillna('nan', inplace=True)
train['ContractTime'].fillna('nan', inplace=True)
```

```python
enc = DictVectorizer()
X_train_categ = enc.fit_transform(train[['LocationNormalized',
                                              'ContractTime']].to_dict('records'))
```

```python
t_f = coo_matrix(text)
t_f2 = coo_matrix(X_train_categ)
print(t_f.shape, t_f2.shape)
train_f = hstack([t_f, t_f2])
```

```python
clf = Ridge(alpha = 1, random_state= 241)
clf.fit(train_f, train['SalaryNormalized'])
```

```python
test_data = pandas.read_csv('salary-test-mini.csv')
```

```python
for i in range(len(test_data['FullDescription'])):
    test_data['FullDescription'][i] = test_data['FullDescription'][i].lower()
test_data['FullDescription'] = test_data['FullDescription'].replace('[^a-zA-Z0-9]',
                                                                    ' ', regex = True)


test_t = full.transform(test_data["FullDescription"])
print(test_t.shape)
```

```python
X_test_categ = enc.transform(test_data[['LocationNormalized',
                                              'ContractTime']].to_dict('records'))
print(X_test_categ.shape)
```

```python
test_f = hstack([test_t, X_test_categ])
```

```python
clf.predict(test_f)
```
