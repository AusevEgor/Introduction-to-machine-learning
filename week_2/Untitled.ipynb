```python
import pandas
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
train = pandas.read_csv('perceptron-train.csv')
test = pandas.read_csv('perceptron-test.csv')
```

```python
x_train = []
y_train = []
for i in range(len(train['F_1'])):
    temp = []
    y_train.append(train['F_1'][i])
    temp.append(train['F_2'][i])
    temp.append(train['F_3'][i])
    x_train.append(temp)
x_train = np.array(x_train)
y_train = np.array(y_train)

x_test = []
y_test = []
for i in range(len(test['F_1'])):
    temp = []
    y_test.append(test['F_1'][i])
    temp.append(test['F_2'][i])
    temp.append(test['F_3'][i])
    x_test.append(temp)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
clf = Perceptron(random_state = 241)
clf.fit(x_train, y_train)
predicts = clf.predict(x_test)
score = accuracy_score(y_test, predicts)
print(score)
```

```python
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)
```

```python
clf_scaler = Perceptron(random_state = 241)
clf_scaler.fit(x_train_scaler, y_train)
predicts_scaler = clf_scaler.predict(x_test_scaler)
score_scaler = accuracy_score(y_test, predicts_scaler)
print(score_scaler)
print(score_scaler - score)
```
