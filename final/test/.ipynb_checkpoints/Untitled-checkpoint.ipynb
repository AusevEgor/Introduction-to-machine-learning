```python
import pandas as pd
import numpy as np
import datetime
from sklearn import model_selection, metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
data = pd.read_csv('features.csv')
data.head()
```

```python
right = max(data.count())
missing_columns = []
for i in range(0,109):
    if data.iloc[:, i].count() != right:
        missing_columns.append(data.columns[i])
print(missing_columns)
data = data.fillna(value=0)
```

```python
right = max(data.count())
missing_columns = []
for i in range(0,109):
    if data.iloc[:, i].count() != right:
        missing_columns.append(data.columns[i])
print(missing_columns)
```

```python
X = data.iloc[:, 2:103]
X.head()
# последние столбцы не берём, потому что это как раз то, что алгоритм знать не должен.
```

```python
y = data['radiant_win']
```

```python
cv = model_selection.KFold(n_splits=5, shuffle=True, random_state=241)
gbc = GradientBoostingClassifier(random_state=241)
params = {'n_estimators': [10, 20, 30]}
gs = model_selection.GridSearchCV(gbc, params, cv=cv, scoring='roc_auc')
```

```python
gs.fit(X, y)
```

```python
gs.best_estimator_
y_pred = gs.predict(X)
gs.score(X, y)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y)
start_time = datetime.datetime.now()
gbc = GradientBoostingClassifier(n_estimators=30, random_state=241)
gbc.fit(x_train, y_train)
print('Time elapsed:', datetime.datetime.now() - start_time)
```

```python
gbc = GradientBoostingClassifier(max_depth=4, learning_rate=0.2, n_estimators=70, random_state=241)
gbc.fit(x_train, y_train)
```

```python
y_pred = gbc.predict_proba(x_test)[:, 1]
```

```python
metrics.roc_auc_score(y_score=y_pred, y_true=y_test)
```

```python
log = LogisticRegression(penalty='l2', solver='lbfgs', random_state=241)
params = {'C': [0.0001, 0.001, 0.01, 0.1]}
gsl = model_selection.GridSearchCV(log, params, cv=cv, scoring='roc_auc')
```

```python
gsl.fit(X, y)
```

```python
gsl.best_estimator_
```

```python
gsl.best_score_
```

```python
X_pick = np.zeros((data.shape[0], 112))
for i, match_id in enumerate(data.index):
    for p in range(5):
        X_pick[i, data.loc[match_id, 'r%d_hero' % (p+1)]-1] = 1
        X_pick[i, data.loc[match_id, 'd%d_hero' % (p+1)]-1] = -1
```

```python
X_pick = pd.DataFrame(X_pick)
x = pd.concat([X, X_pick], axis=1)
```

```python
cv = model_selection.KFold(n_splits=5, shuffle=True, random_state=241)
log = LogisticRegression(penalty='l2', solver='lbfgs', random_state=241)
params = {'C': [0.0001, 0.001, 0.01, 0.1]}
gsl = model_selection.GridSearchCV(log, params, cv=cv, scoring='roc_auc')
```

```python
gsl.fit(x,y)
```
